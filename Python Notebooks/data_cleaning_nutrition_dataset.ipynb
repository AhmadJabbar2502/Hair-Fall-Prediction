```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Data/Nutrition_Dataset.csv')
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
data.columns = [col.replace('_', ' ').title().replace(' ', '_') for col in data.columns]
print(data.columns)
```

```python
data.info()
```

```python
# Rename column food to Food
data = data.rename(columns={' food': 'Food'})
```

```python
# Capital the food names
data['Food'] = data['Food'].str.title()
data.head()
```

```python
data.to_csv('../Data/Cleaned_Nutrition_Dataset.csv', index=False)
```

```python
# 1) Top Foods by Nutrition_Density (bar chart)
import matplotlib.pyplot as plt
import pandas as pd

# settings
top_n = 20

# prepare data
top = data[['Food', 'Nutrition_Density']].dropna().sort_values('Nutrition_Density', ascending=False).head(top_n)

# plot
plt.figure(figsize=(10,6))
plt.barh(top['Food'][::-1], top['Nutrition_Density'][::-1])
plt.xlabel('Nutrition Density')
plt.title(f'Top {top_n} Foods by Nutrition Density')
plt.tight_layout()
plt.show()
```

```python
# 2a) Macronutrient Stacked Bar Chart (Fat, Carbohydrates, Protein) for top foods
import matplotlib.pyplot as plt
import pandas as pd
top_n = 12

cols = ['Fat', 'Carbohydrates', 'Protein']
top = data[['Food'] + cols].dropna().sort_values('Protein', ascending=False).head(top_n).set_index('Food')

# Normalize to 100g proportion if desired (optional)
# top_norm = top.div(top.sum(axis=1), axis=0) * 100

plt.figure(figsize=(12,6))
bottom = pd.Series([0]*len(top), index=top.index)
for c in cols:
    plt.bar(top.index, top[c], bottom=bottom, label=c)
    bottom += top[c]
plt.xticks(rotation=45, ha='right')
plt.ylabel('g per serving (or per 100g)')
plt.title(f'Stacked Macro Composition for Top {top_n} Foods (by Protein)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 3a) Fat Profile Comparison (bar: Saturated vs Unsaturated) for top high-fat foods
import matplotlib.pyplot as plt
import pandas as pd
top_n = 15

# create an "unsaturated" column as monounsaturated + polyunsaturated
data['Unsaturated_Fats'] = data[['Monounsaturated_Fats', 'Polyunsaturated_Fats']].sum(axis=1)

cols = ['Saturated_Fats', 'Unsaturated_Fats']
top = data[['Food'] + cols].dropna().sort_values('Saturated_Fats', ascending=False).head(top_n).set_index('Food')

# plot grouped bars
ax = top.plot(kind='bar', figsize=(12,6))
ax.set_ylabel('g')
ax.set_title(f'Saturated vs Unsaturated Fats (Top {top_n} foods by Saturated Fat)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3b) Fat profile scatter: Saturated_Fats vs Polyunsaturated_Fats
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8,6))
plt.scatter(data['Saturated_Fats'], data['Polyunsaturated_Fats'], alpha=0.6)
plt.xlabel('Saturated Fats (g)')
plt.ylabel('Polyunsaturated Fats (g)')
plt.title('Saturated vs Polyunsaturated Fats')
# annotate top outliers
for i, row in data.nlargest(6, 'Saturated_Fats')[['Food', 'Saturated_Fats', 'Polyunsaturated_Fats']].iterrows():
    plt.annotate(row['Food'], (row['Saturated_Fats'], row['Polyunsaturated_Fats']), textcoords="offset points", xytext=(5,5))
plt.tight_layout()
plt.show()
```

```python
# 7b) Vitamin Radar Chart for a single food (example)
import matplotlib.pyplot as plt
import numpy as np

food_name = data['Food'].iloc[0]  # change index or name as needed
row = data[data['Food'] == food_name].iloc[0]
vit_list = ['Vitamin_A','Vitamin_B1','Vitamin_B11','Vitamin_B12','Vitamin_B2','Vitamin_B3','Vitamin_B5','Vitamin_B6','Vitamin_C','Vitamin_D','Vitamin_E','Vitamin_K']
values = row[vit_list].fillna(0).values
# normalizing for plotting (optional)
max_val = max(values.max(), 1)
values = values / max_val

angles = np.linspace(0, 2 * np.pi, len(vit_list), endpoint=False).tolist()
values = np.concatenate((values, [values[0]]))
angles += [angles[0]]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_thetagrids(np.degrees(angles[:-1]), vit_list)
ax.set_title(f'Vitamin Profile (normalized) — {food_name}')
plt.tight_layout()
plt.show()
```

```python
# 13) PCA and KMeans clustering (visualize foods grouped by nutrition profile)
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# choose numeric columns and drop NA
numeric = data.select_dtypes(include=['number']).drop(columns=['Nutrition_Density'], errors='ignore')  # exclude density if desired
numeric = numeric.dropna()
# keep same index alignment for labels
idx = numeric.index
labels = data.loc[idx, 'Food']

# scale, PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric)
pca = PCA(n_components=2)
pc = pca.fit_transform(X_scaled)

# kmeans
k = 4
km = KMeans(n_clusters=k, random_state=42)
clusters = km.fit_predict(X_scaled)

plt.figure(figsize=(10,7))
scatter = plt.scatter(pc[:,0], pc[:,1], c=clusters, cmap='tab10', alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Nutrition Profiles with KMeans clusters')
# annotate some points
for i in range(len(labels)):
    if i % max(1, len(labels)//15) == 0:  # annotate a subset for readability
        plt.annotate(labels.iloc[i], (pc[i,0], pc[i,1]), fontsize=8)
plt.tight_layout()
plt.show()

# optional: cluster centers in PCA space
centers = km.cluster_centers_
centers_pca = pca.transform(centers)
plt.figure(figsize=(8,5))
plt.scatter(pc[:,0], pc[:,1], c=clusters, cmap='tab10', alpha=0.6)
plt.scatter(centers_pca[:,0], centers_pca[:,1], marker='X', s=200, c='black')
plt.title('Clusters and cluster centers (PCA projection)')
plt.tight_layout()
plt.show()
```

```python
# 14) (Optional) KMeans cluster summary table: top features per cluster
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

numeric_cols = data.select_dtypes(include=['number']).columns.tolist()
X = data[numeric_cols].dropna()
idx = X.index
scaler = StandardScaler()
X_s = scaler.fit_transform(X)

k = 4
km = KMeans(n_clusters=k, random_state=42)
clusters = km.fit_predict(X_s)
clustered = data.loc[idx].copy()
clustered['cluster'] = clusters

# compute mean per cluster
cluster_means = clustered.groupby('cluster')[numeric_cols].mean()
display(cluster_means.T)  # use display in notebooks
```
